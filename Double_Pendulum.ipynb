## Introduction 
The double pendulum is widely used in science to study chaos and states transitions. Even tough the movement is described as chaotic, we know there is a certain predictability with specific double pendulum parameters. For example, there is a normal mode, where both pendulums move side to side in an anti-parallel motion. Another instance is when both are moving side to side synchronously (see Figure 1.1 and 1.2). This makes us wonder if there is a certain predictability than can be discovered for small angles (theta 1 and theta 2) that can be manipulated by changing the ratio of lengths. How does the ratio of lengths of the pendulums affect the patterns (if there are any) of the movements of the pendulums for small angles only meaning for such angle that the second pendullum does not loop around the first one? In order to analyze the relation between both these paramaters and the motion of the system, we will simulate a double pendulum with all the others paramaters being fixed. Namely, the gravitational acceleration, the masses, the initial angular velocities of both pendulums and the initial angles of both pendulums will all have fixed values which will create a double pendulum moving without the second pendulum looping aroung the first (which is what we consider as small angle movement in this report). We decided to only analyse this type of movement since the chaotic movement that includes loops has not been proven yet to have a certain predictability (see the figure 2.1 for a reference of the model which will be simulated). 

We believe that a bigger ratio of lengths (rod 1 over rod 2) will result in a smaller range of movement done by the second pendullum relative to the first pendulum. This will be demonstrated with a graph of theta 1 and theta 2 as variables. The reasoning behind it is that the longer rod 2 is in comparison to rod 1 the more effect it will have on the movement of pendulum 1. Therefore, as pendulum 2 is the pendulum that gives the chaotic property to this model, when it swings in an unpredictable motion it will "pull" on the first pendulum and make it have a bigger range of motion. Thus, as pendulum 2 range of motion is dependent on pendulum 1 it will then also see its range increasing by the increase in the range of motion of pendulum 1.

## Theory of the model
Our objective is to simulate a double pendulum, such as the one represented in figure 2.1. In order to achieve our intended objective, we need a model that can simulate the motion of a double pendulum. This type of model is called OOP: object-oriented programming. It is based on the concept of objects, which contain data and code and that can access and modify the data fields of itself.
In addtion, we will need a precise model which will make the double pendulum system respect the law of Conservation of Energy. This will not be possible using Euler's method, because it is only first order convergent. Consequently, the approximation the method will make for each time step will be very poor, and would require a too small time step in order to have a serious accuracy. Instead, a more appropriate method would be the Euler-Lagrange method. This method provides better accuracy for our purposes. We therefore made our simulation of a double pendulum system using the Lagrangian equations. First of all, the equations for the Potential and Kinetic energies that we used were respectively:

$ PE=m_1*g*y_1+m_2*g*y_2 $                                                     

$ KE=*m_1*(L_1 * w_1)^2 +*m_2 *((L_1 * w_1)^2 + (L_2 *w_2)^2 + 2 * L_1 * L_2 * w_1 * w_2 * cos(_1 - _2)) $

where g is the gravitational acceleration, $m_1$ and $m_2$ are the masses of each ball, with $w_1$ and $w_2$ being the angular velocities, $L_1$ and $L_2$ are the lengths of each rod and $_1$ and $_2$ are the angles of the first and second part of the double pendulum, respectively. 
Furthermore, we needed the equations for the declination angles  for each part of the double pendulum, namely:


$_1 = (L_2 / L_1) * (m_2 / (m_1 + m_2)) *cos(_1 - _2)$

$_2 =(L_1 / L_2) *cos(_1 - _2)$

Both these values will be used to find the two angular accelerations $g_1$ and $g_2$:

$g_1=(f_1 - _1 * f_2) (1 - _1 * _2)$

$g_2=(f_2 - _2 * f_1)  (1 - _1 * _2)$

with $f_1$ and $f_2$ being, respectively,


$f_1=-(L_2 / L_1) * (m_2 / (m_1 + m_2)) * (w_2^2) *sin(t_1 - t_2) -(g / L_1) *sin(_1)$

$f_2=(L_1 / L_2) * (w_1^2) *sin(_1 - _2) - (g / L_2) *sin(_2)$

To achieve our objective, there will be 4 different phases. In the first phase, we will set the initial parameters for the simulation, namely the parameters needed to compute the equations above. To continue, in the second phase, the potential energy and the kinematic energy will be computed using the respective Lagrange equations. Also, the new angles and the new angular velocities will be determined by using the Runge-Kutta method. More precisely, the fourth order Range-Kutta method will be used, as it is more accurate than the lower order ones. This is an effective method to solve initial-value problems of differential equations. It takes a weighted average of the slopes at a large number of points and is therefore also more accurate than Euler's method. The equations used for the double pendulum system are 
	
$k_1=f(y)$

$k_2=f(y+dt*k_1/2)$

$k_3=f(y+dt*k_2/2)$

$k_4=f(y+dt*k_3)$

with $k_1$ being the slope at the beginning of the time step. Using the slope $k_1$ to step halfway through the time step, $k_2$ is an estimate of the slope at the midpoint. In a similar process, $k_3$ is another estimate of the slope at the midpoint. Then, using the slope $k_3$ to step to the end of the time step, $k_4$ will provide an estimate of the slope at the end point. This method is clearly more accurate than Euler's method. In the third phase, the double pendulum will be simulated. In fact, a loop will be created in order to update and save all the values of angles, angular velocities and energies over a period of time. 
Finally, to obtain the positions of each ball at each moment in the simulation, the lengths of the pendulums and the deflection angles will be used with trigonometry such as the cos function and the sin function, depending on if we want to find the vertical or the horizontal position. That is, we will use the following pendulum equations:

$ x_{1}=L_1*sin(_1) $

$ x_{2}=L_1*sin(_1) + L_2*sin(_2) $

$y_{1}=-L_1*cos(_1)$

$ y_{2} = -L_1*cos(_1)-L_2*cos(_2) $

With all this done, we will be able to create graphs comparing the parameters that interest us in order to find a solution to our problem.

## Description of the numerical method
The simulation of the double pendulums was accomplished using numerical methods.
Our model of the double pendulums uses time steps to record the angles and the angular velocities of the two masses at each time step by using the Euler-Lagrange equations and their derivatives. Once obtaining the angles and angular velocities at each time step, we were able to then calculate the horizontal and vertical positions of the two masses by using trigonometry with the angles and the lengths of the rods and with the first pivot being the point of origin at (0,0). We were also able to calculate the kinetic and potential energies at each time step by using the masses, the speeds and the vertical positions of the pendulums.


## Setup for the code (Phase 1)
In the first phase of the code, we created a class that would hold all of the parameters of the double pendulum. To do this, we defined the double pendulum as an object that contained the gravity, the masses, the angles, the angular velocities, the lengths of the rods, the amount of time between each step (t), as well as lists that would contain each instances of the aforementionned angles, angular velocities, potential and kinetic energies, as well as a list representing the time by adding one t at each instance.

## Creating the functions (Phase 2)
In the second phase of the code, we firstly created a function to determine the potential energy by computing the heights of the pendulums (using trigonometry), with (0,0) being the pivot of the first pendulum, and by computing and adding the potential energies of each pendulum using the formula for the potential energy. Then, we computed the kinetic energy of the system by calculating the individual kinetic energies of the pendulums using the angular velocities, the lengths and the masses of the pendulums, and then adding them both together to find the total kinetic energy of the system. Finally, the time step function uses the Lagrange function and the Runge-Kutta Method to determine the new angles and the angular velocities at the next time step.

## Simulating the double pendulum (Phase 3)
In the third phase of the code, we simulated the motion of the double pendulum by creating a while loop in which the time_step function and the lagrange_rhs would run for each time step, which would thus update and accumulate in lists all the values of the angles and angular velocities as well as the potential and kinetic energies, over a period of time T.

## Extracting the positions of the double pendulum (Phase 4)
Once we have run the initial block of code and simulated the motion of an instance of the class of the double pendulum, we can then figure out the lists for the vertical and horizontal positions of both pendulums by using trigonometry with the lists of angles of deflection and the lengths of the pendulums, by comparing them with the first pendulum's pivot as the origin (0,0) and the reference point.

##Code

```python
#Import libraries which will be necessary to simulate a double pendulum
import math
import numpy as np
import pandas as pd


class DoublePendulumLagrangian:
  #Constants which apply for all instances
  dt=0.01
  T=40
  steps=(T/dt)

  def __init__(self, g, m_1, m_2, t_1, t_2, w_1, w_2, L_1, L_2,dt,list_t_1,list_t_2,list_w_1,list_w_2,time_index,list_potential_energy,list_kinetic_energy):
        #Initiliaze the parameters necessary to construct a double pendulum simulation.
        #The Lagrange eqautions will be used.
        #Ball 1 is attached to a fixed pivot, while the Ball 2 is attached to Ball 1.
     
        self.g = g           #The gravitational acceleration
        self.m_1 = m_1       #The mass of Ball 1   
        self.m_2 = m_2       #The mass of Ball 2
        self.t_1 = t_1       #The initial angle of Ball 1
        self.t_2 = t_2       #The initial angle of Ball 2
        self.w_1 = w_1       #The initial angular velocity of Ball 1 
        self.w_2 = w_2       #The initial angular velocity of Ball 2         
        self.L_1 = L_1       #The length of the rod linking Ball 1 to the fixed pivot
        self.L_2 = L_2       #The length of the rod linking Ball 2 to Ball 1
        self.dt=dt           #The time interval between each step 
        self.list_t_1=[]     #Initialization of the list which will save the t_1 at each step
        self.list_t_2=[]     #Initialization of the list which will save the t_2 at each step
        self.list_w_1=[]     #Initialization of the list which will save the w_1 at each step
        self.list_w_2=[]     #Initialization of the list which will save the w_2 at each step
        self.time_index=[0,] #Initialization of the list which will save the time at each step
        self.list_potential_energy=[] #Initialization of the list which will save the potential energy at each step
        self.list_kinetic_energy=[] #Initialization of the list which will save the kinetic energy at each step
        
  def potential_energy(self):

        #Compute the potential energy of the double pendulum system
        m_1 = self.m_1
        t_1 = self.t_1
        L_1 = self.L_1
        m_2 = self.m_2
        t_2 = self.t_2
        L_2 = self.L_2
        g = self.g

        #Calculate the height of each Ball
        y1 = -L_1 * math.cos(t_1)
        y2 = y1 - L_2 * math.cos(t_2)
        return m_1 * g * y1 + m_2 * g * y2
    
  def kinetic_energy(self):

        #Compute the kinetic energy of the double pendulum system
        m_1 = self.m_1
        t_1 = self.t_1
        w_1 = self.w_1
        L_1 = self.L_1
        m_2 = self.m_2
        t_2 = self.t_2
        w_2 = self.w_2
        L_2 = self.L_2

        # compute the kinetic energy of each ball
        K1 = 0.5 * m_1 * (L_1 * w_1)**2
        K2 = 0.5 * m_2 * ((L_1 * w_1)**2 + (L_2 * w_2)**2 +
                         2 * L_1 * L_2 * w_1 * w_2 * math.cos(t_1 - t_2))
        return K1 + K2

  def lagrange_rhs(self, t_1, t_2, w_1, w_2):

        #Compute the right-hand side of the Lagrange equations for the double pendulum and return it as an array.     
        import pandas as pd
        import numpy as np
        m_1 = self.m_1
        L_1 = self.L_1
        m_2 = self.m_2
        L_2 = self.L_2
        g = self.g

        a_1 = (L_2 / L_1) * (m_2 / (m_1 + m_2)) * math.cos(t_1 - t_2)   #a_1 and a_2 are the deflection angles for each of the two parts of the pendulum
        a_2 = (L_1 / L_2) * math.cos(t_1 - t_2)

        f_1 = -(L_2 / L_1) * (m_2 / (m_1 + m_2)) * (w_2**2) * math.sin(t_1 - t_2) - (g / L_1) * math.sin(t_1)
        f_2 = (L_1 / L_2) * (w_1**2) * math.sin(t_1 - t_2) - (g / L_2) * math.sin(t_2)

        g1 = (f_1 - a_1 * f_2) / (1 - a_1 * a_2)
        g2 = (f_2 - a_2 * f_1) / (1 - a_1 * a_2)
        return np.array([w_1, w_2, g1, g2])

  def time_step(self, dt):

        #Advance one time step using the classical Runge-Kutta method
        m_1 = self.m_1
        t_1 = self.t_1
        w_1 = self.w_1
        L_1 = self.L_1
        m_2 = self.m_2
        t_2 = self.t_2
        w_2 = self.w_2
        L_2 = self.L_2
        dt=self.dt

        # y is an array with the generalized coordinates (angles and angular velocities)
        y = np.array([t_1, t_2, w_1, w_2])  

        # compute the Runge-Kutta constants
        k1 = self.lagrange_rhs(*y)
        k2 = self.lagrange_rhs(*(y + dt * k1 / 2))
        k3 = self.lagrange_rhs(*(y + dt * k2 / 2))
        k4 = self.lagrange_rhs(*(y + dt * k3))

        # compute the Runge-Kutta right-hand side
        R = 1.0 / 6.0 * dt * (k1 + 2.0 * k2 + 2.0 * k3 + k4)

        # update the angles and angular velocities
        self.t_1 += R[0]
        self.t_2 += R[1]
        self.w_1 += R[2]
        self.w_2 += R[3]
        self.list_t_1.append(t_1)
        self.list_t_2.append(t_2)
        self.list_w_1.append(w_1)
        self.list_w_2.append(w_2)

        #update the kinetic and potential energies of the system
        self.list_potential_energy.append(self.potential_energy())
        self.list_kinetic_energy.append(self.kinetic_energy())

  def simulate(self):
    # This is the main loop for which all the above functions are running until the maximum time has been reached
    
    self.potential_energy()
    self.kinetic_energy()
    i=0

    dt=0.01
    T=40
    steps=(T/dt)
    while i<steps:
      self.lagrange_rhs(self.t_1,self.t_2,self.w_1,self.w_2)
      self.time_step(self.dt)
      i=i+1
    i=0
    while i<(steps-1): #This is done so that the time_index has the same shape as the other parameters,
    #as it was already given an initial value.
      self.time_index.append(self.time_index[-1]+self.dt)
      i=i+1
```

```python
## Define Double Pendulum instance DP with set initial parameters
DP=DoublePendulumLagrangian(
    9.8,
    1,
    1,
    0.5,
    0.5,
    0,
    0,
    0.5,
    1,
    0.01,
    [],
    [],
    [],
    [],
    [],
    [],
    [],)
```

```python
#Simulate the double pendulum system
DP.simulate()
```

```python
dt=0.01
T=40
steps=(T/dt)

#Create the lists which save the x-positions for each of the pendulums, based on the length of the respective rod
#and on the respective angle at each time step
list_x_1=[]
i=0
while i<steps:
  list_x_1.append(DP.L_1*math.sin(DP.list_t_1[i]))
  i=i+1

list_x_2=[]
i=0
while i<steps:
  list_x_2.append(DP.L_2*math.sin(DP.list_t_2[i])+list_x_1[i])
  i=i+1

#Create the lists which save the y-positions for each of the pendulums, based on the length of the respective rod
#and on the respective angle at each time step
list_y_1=[]
i=0
while i<steps:
  list_y_1.append(-(DP.L_1*math.cos(DP.list_t_1[i])))
  i=i+1

list_y_2=[]
i=0
while i<steps:
  list_y_2.append((list_y_1[i])-(DP.L_2*math.cos(DP.list_t_2[i])))
  i=i+1

  
```

###Discussion: 
Our results indicate that, as predicted by our hypothesis, the range of motion of pendulum 2 according to pendulum 1 is bigger when the ratio of L1:L2 is smaller. And on the contrary, the movement in both angles is bigger when the ratio is smaller. For all graphs that represent the relationship of theta 1 and theta 2 we get an irregular shape that visibly alters in area. That area can be related to the range of motion of both angles and thus the range of motion of the second pendulum. For the graph representing a pendulum of length ratio 2:1(green color graph) when theta 1 is equal to 0, the width of the graph shape goes roughly from -0.20rad to 0.20rad and that gives a width of 0.40 rad which is around half of what the width of the shape in red color is. This makes sense because the shorter L2 is compared to L1, the less influence pendulum 2 movement has on pendulum 1 and therefore the overall movement is more rigid. On the contrary, the longer L2 is compared to L1, the more influence its movement has on the movement of pendulum 1 and as pendulum 2 is dependent on pendulum 1, the movement range of pendulum 2 increases even more. That proves that our hypothesis was right and that the bigger the ratio of L1:L2 is the smaller the range of motion of the two angles is for the graph that represents the movement of the first pendulum, it makes sense that it is in a parabola shape as it is fixed to the point(0,0) that does not move. The graph representing the total energy of the system shows that the system's energy stays constant. This allows us to confidently determine that there is a certain predictability in the motion of the double pendulum as long as its angles are small enough for the second pendulum to not loop on the first one.
